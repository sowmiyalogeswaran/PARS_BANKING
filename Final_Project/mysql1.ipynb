```python
import pandas as pd
import mysql.connector

# Load the CSV file
df = pd.read_csv("simplified_dataset.csv")
```

```python
df
```

```python
# Connect to MySQL
my_connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='3008',
    auth_plugin='mysql_native_password'
)
my_cursor = my_connection.cursor()


```

```python
# Creating the database 
my_cursor.execute("CREATE DATABASE IF NOT EXISTS pars_banking")

# Switch to new database
my_cursor.execute("USE pars_banking")
```

```python
# Create table (if not already created)
create_table_query = '''
CREATE TABLE IF NOT EXISTS simplified_data (
    Customer_Id VARCHAR(255),
    Age INT,
    Monthly_Income FLOAT,
    Credit_Score INT,
    Credit_Score_Band VARCHAR(50),
    Loan_Amount FLOAT,
    Interest_Rate FLOAT,
    Loan_Term INT,
    Loan_Type VARCHAR(50),
    Debt_Income_Ratio FLOAT,
    Repayment_Status INT,
    Transaction_Id VARCHAR(255),
    Gender VARCHAR(10),
    Transaction_Amount FLOAT,
    Transaction_Type VARCHAR(50),
    Transaction_Date DATE,
    Transaction_Year INT,
    Transaction_Month INT,
    Product_Id VARCHAR(255),
    Product_Name VARCHAR(255),
    Interaction_Date DATE,
    Interaction_Year INT,
    Interaction_Day_Name VARCHAR(20),
    Interaction_Type VARCHAR(50),
    Transaction_Frequency INT
);
'''
my_cursor.execute(create_table_query)
```

```python
# Insert data from DataFrame into MySQL table
for row in df.itertuples(index=False, name=None):
    my_cursor.execute('''
    INSERT INTO simplified_data (
        Customer_Id, Age, Monthly_Income, Credit_Score, Credit_Score_Band, Loan_Amount, 
        Interest_Rate, Loan_Term, Loan_Type, Debt_Income_Ratio, Repayment_Status, 
        Transaction_Id, Gender, Transaction_Amount, Transaction_Type, Transaction_Date, 
        Transaction_Year, Transaction_Month, Product_Id, Product_Name, Interaction_Date, 
        Interaction_Year, Interaction_Day_Name, Interaction_Type, Transaction_Frequency
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s)
    ''', row)

# Commit the transaction to save the changes
my_connection.commit()

# Close the cursor and connection
my_cursor.close()
my_connection.close()
```
